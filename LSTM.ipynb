```python
import numpy as np
import pandas as pd
```

```python
#Calculate log returns
data = pd.read_excel('N225.xlsx').dropna(axis=0)
data = data.drop(['Volume', 'Open', 'High', 'Low', 'Close'], axis=1)
data['Log_return'] = np.log(data['Adj Close']).diff()
```

```python
#Calculating rolling volatility and mean return using window size = 120
def get_rolling_vol(vol_wind,lst):
    vol_lst = []
    for i in range(len(lst)):
        if i<vol_wind:
            vol_lst.append(0)
        else:
            vol_lst.append(np.std(lst[i-vol_wind:i]))
    return np.array(vol_lst)

data['120-day vol'] = get_rolling_vol(120,data['Log_return'])
data

def get_rolling_mean(mean_wind,lst):
    mean_lst = []
    for i in range(len(lst)):
        if i<mean_wind:
            mean_lst.append(0)
        else:
            mean_lst.append(np.mean(lst[i-mean_wind:i]))
    return np.array(mean_lst)

data['120-day mean'] = get_rolling_mean(120,data['Log_return'])
```

```python
data = data[~(data == 0).any(axis=1)]
thresh = np.median(data['120-day vol'].values)
data['regime'] = np.ones(len(data))
for i in range(len(data)):
    if data.iloc[i,3] >= thresh:
        data.iloc[i,5] = 1
    else:
        data.iloc[i,5] = 2
```
